```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on = 'title') #merging all columns
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']] #taking columns that are needed
```

```python
movies.head(1)
```

```python
movies.dropna(inplace = True) 
#dropna removes the row that have empty column
#inplace changes the actual csv file
```

```python
movies.duplicated().sum() #checking
```

```python
import ast #converts string into list
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name']) #appending every word of obj in the lis L
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert) #string -> obj -> list
```

```python
movies['keywords'] = movies['keywords'].apply(convert) #string  -> list 
```

```python
movies.head(1) #checking the result of genres and keywords
```

```python
def convertCast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3: #first 3 names are needed
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convertCast) #first 3 name -> list
```

```python
movies.head(1)
```

```python
def findDirector(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(findDirector)
```

```python
movies.head(1)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #spliting every word and create a list
```

```python
movies['overview'][0]
```

```python
movies.head()
```

```python
#removing spaces in the columns
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#adding a new columns name tag which is the combination of all columns added below
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']] #making new dataframe haveing necessary columns only
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x)) #converving string
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower()) #lower casing
```

```python
new_df.head()
```

```python
!pip3 install nltk
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
#making feature(column) for 5000 words and removing english unnecessary words(in, to, from, ...)
cv = CountVectorizer(max_features = 5000, stop_words = 'english') 
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() #vectorizing (2D)
```

```python
vectors
```

```python
cv.get_feature_names_out() #checking features(column)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors) #calculating distances
```

```python
similarity[0] #distance grom first movie to others
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] #indexing
    movie_name = sorted(list(enumerate(similarity[movie_index])),reverse = True, key = lambda x:x[1])[1:6] #holding the index -> converign into list -> sorted first 5 movies based on distance(2nd column)

    for i in movie_name:
        print(new_df.iloc[i[0]].title) #printing the movies
```

```python
recommend('Avatar')
```
